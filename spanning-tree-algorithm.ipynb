## Graph

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7', 'HA', 'HB', 'HC'])
G.add_edges_from([('S1', 'S2'), ('S1', 'S4'), ('S1', 'S5'), ('S1', 'S6'), 
                  ('S1', 'S7'), ('S6', 'S4'), ('S5', 'S7'), ('S3', 'S2'), 
                  ('S3', 'S5'), ('HA', 'S3'), ('HB', 'S7'), ('HC', 'S4')])
nx.draw_networkx(G, with_labels=True)
plt.show()
```

## STA (Spinning Tree Algorithm)

```python
T = nx.minimum_spanning_tree(G)
nx.draw_networkx(T, with_labels=True)
plt.show()
```
